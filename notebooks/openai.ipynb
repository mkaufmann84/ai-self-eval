```python
import os
from dotenv import load_dotenv

load_dotenv("../.env.local")

#var = os.getenv('OPENAI_API_KEY')

from openai import OpenAI
client = OpenAI()

completion = client.chat.completions.create(
  model="gpt-4o",
  messages=[
    {"role": "user", "content": """How would you define what the "rules" of something are without a circular defination in a single sentence?"""}
  ],
  logprobs=True,
  top_logprobs=5,
  n=1,
  temperature=0,
  top_p=1,
)

```

```python
import pprint
```

```python
from openai.types.chat import ChatCompletion
import math
def pretty_completion(completion: ChatCompletion):
    for choice in completion["choices"]:
        for token in choice["logprobs"]["content"]:
            del token["bytes"]
            token["prob"] = math.exp(token["logprob"])
            del token["logprob"]
            for top_logprob in token["top_logprobs"]:
                del top_logprob["bytes"]
                top_logprob["prob"] = math.exp(top_logprob["logprob"])
                del top_logprob["logprob"]
    pprint.pprint(completion["choices"])
```
